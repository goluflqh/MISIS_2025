# 1. Что такое Веб? (Frontend vs Backend)

Любой сайт — это ресторан.

1.  **Frontend (Зал ресторана):**
    * Это то, что видит пользователь в браузере.
    * Кнопки, красивые шрифты, анимации.

    * *В нашем курсе мы будем использовать готовые шаблоны, чтобы не учить CSS с нуля.*

2.  **Backend (Кухня):**
    * Это то, что происходит на сервере.
    * Обработка заказов, работа с базой данных, логика.
    * Здесь поваром работает наш **Python**.

## 2. Как общаются компьютеры? (HTTP)

Весь интернет работает по принципу **"Вопрос — Ответ"**.

1.  **Client (Браузер):** "Привет, сервер! Дай мне главную страницу сайта `google.com`". Это называется **Request** (Запрос).
2.  **Server (Наш код):** "Привет! Вот HTML-код этой страницы". Это называется **Response** (Ответ).


Frontend👩‍🎨 (пафосный чувак с смузи): Всё, что видит пользователь — кнопки, формы, котиков на фоне.  
Backend👨‍💻 (грязный подвал с замученными кодерами): Всё, что происходит "за сценой" — обработка запросов, сохранение данных, отправка ваших постов на сервер

# Разделяй и Властвуй (Templates) 📑

Посмотрите на код из прошлой части:

Представьте, что вам нужно сверстать сложную страницу с таблицей расписания, кнопками и меню. Ваша строка `return` растянется на 500 символов. В ней будет невозможно найти ошибку. А если дизайнер захочет поменять цвет кнопки, ему придется лезть в ваш Python-код.

Это нарушает главный принцип разработки: **Separation of Concerns (Разделение ответственности)**.

* **Python** должен заниматься логикой (считать, искать в базе).
* **HTML** должен отвечать за внешний вид.

````python
from werkzeug.serving import run_simple
from flask import Flask
from flask import request

app = Flask(__name__)

@app.route("/")
def home():
    return "<h1>Добро пожаловать в мое первое Flask-приложение!</h1>"

@app.route("/hello")
def hello():
    return "Привет, друг! Это страничка /hello."

@app.route("/user/<name>")
def user(name):
    return f"Приятно познакомиться, {name.title()}!"

@app.route("/html")
def html_example():
    return """
    <html>
        <head>
            <title>Flask App</title>
        </head>
        <body>
            <h1>Привет из Flask!</h1>
            <p>Это простая HTML-страница, которая оживает благодаря магии Python и HTML.</p>
        </body>
    </html>
    """
@app.route("/form", methods=["GET", "POST"])
def form_example():
    if request.method == "POST":
        name = request.form.get("name")
        return f"<h1>Привет, {name.title()}! Вот наше Flask-приложение делает чудеса 🪄</h1>"

    return """
    <html>
        <body>
            <h1>Введите своё имя</h1>
            <form method="POST">
                <label for="name">Имя:</label>
                <input type="text" id="name" name="name">
                <input type="submit" value="Отправить">
            </form>
        </body>
    </html>
    """

run_simple('localhost', 5047, app, use_reloader=False)
````

### Решение: Шаблонизатор (Template Engine)

Flask использует встроенный движок **Jinja2**.
Он позволяет нам писать обычные HTML-файлы, но вставлять в них "дырки" (placeholder), в которые Python будет заливать данные.

**Как это работает:**
1. Мы создаем папку `templates` рядом с нашим скриптом. Flask ищет HTML **только** там.
2. В Python мы используем функцию `render_template('index.html')`.

## 2. Магия Jinja2: Как скрестить Питона и HTML?

HTML — это статический текст. Он не умеет делать циклы `for` или проверки `if`.
Jinja2 добавляет в HTML суперсилы с помощью специальных скобочек.

Запомните два типа скобок:

1.  `{{ переменная }}` — **Напечатать значение**.
    * Если Python передаст `name="Иван"`, то `<h1>Привет, {{ name }}!</h1>` превратится в `<h1>Привет, Иван!</h1>`.

2.  `{% команда %}` — **Логический блок**.
    * Позволяет использовать `if`, `else` и `for` прямо внутри HTML.

Пример (создаем список студентов):
```html
<ul>
  {% for student in students_list %}
      <li>{{ student }}</li>
  {% endfor %}
</ul>
```

#  Почтальон стучится дважды (GET vs POST) 📬

Чтобы загрузить файл, нам нужно разобраться, как браузер общается с сервером.
Существует два главных метода (глагола) в протоколе HTTP:

1.  **GET (Дай):**
    * Используется, когда мы заходим на сайт по ссылке.
    * Мы говорим серверу: *"Эй, покажи мне страницу /admin"*.
    * Параметры передаются в адресе: `google.com/?q=котики`.

2.  **POST (На, держи):**
    * Используется, когда мы заполняем формы (логин, регистрация, загрузка файла).
    * Мы говорим серверу: *"Эй, я принес тебе данные, обработай их"*.
    * Данные спрятаны внутри запроса (в "конверте").

##### Отличие POST от GET 

И POST и GET - это HTTP-методы.

Давайте рассмотрим что  из себя представляют HTTP-методы GET/POST/PUT/DELETE и другие, для чего они были придуманы и как их использовать в соответствии с REST.

### HTTP

#### METHOD URI HTTP/VERSION 

где METHOD — это как раз метод HTTP-запроса, URI — идентификатор ресурса, VERSION — версия протокола 

Заголовки — это набор пар имя-значение, разделенных двоеточием. В заголовках передается различная служебная информация: кодировка сообщения, название и версия браузера, адрес, с которого пришел клиент (Referrer) и так далее.

Тело сообщения — это, собственно, передаваемые данные. В ответе передаваемыми данными, как правило, является html-страница, которую запросил браузер, а в запросе, например, в теле сообщения передается содержимое файлов, загружаемых на сервер. Но как правило, тело сообщения в запросе вообще отсутствует.


#### Пример HTTP-взаимодействия


Рассмотрим пример

Запрос:


```

```

#### Ресурсы и методы

параметр URI - Uniform Resource Identifier — единообразный идентификатор ресурса. Ресурс — это, как правило, файл на сервере (пример URI в данном случае '/styles.css'), но вообще ресурсом может являться и какой-либо абстрактный объект ('/blogs/webdev/' — указывает на блок «Веб-разработка», а не на конкретный файл).

Тип HTTP-запроса (также называемый HTTP-метод) указывает серверу на то, какое действие мы хотим произвести с ресурсом. Изначально (в начале 90-х) предполагалось, что клиент может хотеть от ресурса только одно — получить его, однако сейчас по протоколу HTTP можно создавать посты, редактировать профиль, удалять сообщения и многое другое. И эти действия сложно объединить термином «получение».

Для разграничения действий с ресурсами на уровне HTTP-методов и были придуманы следующие варианты:

`
GET — получение ресурса
POST — создание ресурса
PUT — обновление ресурса
DELETE — удаление ресурса
`

спецификация HTTP не обязывает сервер понимать все методы (которых на самом деле гораздо больше, чем 4) — обязателен только GET, а также не указывает серверу, что он должен делать при получении запроса с тем или иным методом. А это значит, что сервер в ответ на запрос DELETE /index.php HTTP/1.1 не обязан удалять страницу index.php на сервере, так же как на запрос GET /index.php HTTP/1.1 не обязан возвращать вам страницу index.php, он может ее удалять, например

#### REST

REST (REpresentational State Transfer)

 REST охватывает более широкую область, нежели HTTP — его можно применять и в других сетях с другими протоколами. REST описывает принципы взаимодействия клиента и сервера, основанные на понятиях «ресурса» и «глагола» (можно понимать их как подлежащее и сказуемое). В случае HTTP ресурс определяется своим URI, а глагол — это HTTP-метод.

REST предлагает отказаться от использования одинаковых URI для разных ресурсов (то есть адреса двух разных статей вроде /index.php?article_id=10 и /index.php?article_id=20 — это не REST-way) и использовать разные HTTP-методы для разных действий. То есть веб-приложение, написанное с использованием REST подхода будет удалять ресурс при обращении к нему с HTTP-методом DELETE (разумеется, это не значит, что надо давать возможность удалить всё и вся, но любой запрос на удаление в приложении должен использовать HTTP-метод DELETE).

REST дает программистам возможность писать стандартизованные и чуть более красивые веб-приложения, чем раньше. Используя REST, URI для добавления нового юзера будет не /user.php?action=create (метод GET/POST), а просто /user.php (метод строго POST).

В итоге, совместив имеющуюся спецификацию HTTP и REST-подход наконец-то обретают смысл различные HTTP-методы. GET — возвращает ресурс, POST — создает новый, PUT — обновляет существующий, DELETE — удаляет.

### Резюме:

Общего между ними то что они работают одинаково. Разницы между ними технически никакой. А вот идеологические различия есть.

GET запрос используется чтобы получить данные а POST чтобы отправить.

GET - это чтобы получить определенную страницу в определенном виде ( сортировка, текущая страница в блоге, строка поиска и т.п. ).

POST - для оправки данных которые не влияют на отображение страницы, в том плане что эти данные влияют только на результат выполнения скрипта ( логины, пароли, номера кредиток, сообщения и т.п. ).

````python
import os
from flask import Flask, render_template, request
# Импортируем наш класс парсера (файл schedule_parser.py должен лежать рядом!)
from schedule_parser import ScheduleParser

app = Flask(__name__)
UPLOAD_FOLDER = 'data'
if not os.path.exists(UPLOAD_FOLDER): os.makedirs(UPLOAD_FOLDER)

@app.route('/', methods=['GET', 'POST'])
def index():
    # Если просто зашли на сайт - показываем форму загрузки
    if request.method == 'GET':
        return render_template('admin.html') # Используем старый шаблон с формой
    
    # Если отправили файл
    if 'schedule_file' not in request.files: return "Нет файла"
    file = request.files['schedule_file']
    if file.filename == '': return "Файл не выбран"

    if file:
        # 1. Сохраняем
        path = os.path.join(UPLOAD_FOLDER, 'uploaded_schedule.xlsx')
        file.save(path)
        
        # 2. ПАРСИМ (Интеграция!)
        print("⚙️ Запускаю парсер...")
        parser = ScheduleParser(path)
        parsed_data = parser.parse()
        
        # 3. Отдаем данные в новый шаблон
        # Обратите внимание: мы передаем словарь parsed_data в переменную schedule
        return render_template('preview.html', schedule=parsed_data)

if __name__ == '__main__':
    # Запускаем!
    app.run(port=5000, debug=True, use_reloader=False)
````

REST API — это интерфейс, который позволяет клиентам (например, приложениям, другим серверам или вам лично через браузер) взаимодействовать с вашим сервером 🔗.
Представь: клиент посылает запрос в виде "Эй, сервер, дай перечень книг по программированию!", и сервер отвечает "Окей, вот JSON с книгами."
Всё максимально структурировано, модно, молодёжно. 

Почему JSON?


JSON — это как Excel для серверов: 

Удобен для чтения человеком и машиной  
Идеален для передачи информации.

Давайте сделаем свой первый REST API.



Давайте начнём с простого примера API, которое возвращает JSON с информацией.



````python
from werkzeug.serving import run_simple
from flask import Flask
from flask import request
from flask import render_template
from flask import jsonify


app = Flask(__name__)

@app.route("/")
def home():
    return "<h1>Добро пожаловать в мое первое Flask-приложение!</h1>"

@app.route("/api/data")
def api_data():
    return jsonify({
        "id": 1,
        "name": "Flask API Example",
        "description": "Это пример JSON-ответа из Flask API."
    })


run_simple('localhost', 5066, app, use_reloader=False)
````

### 4.4 Более сложный API с POST-запросом


````python
Хотим круче? Добавим POST-запрос. Это позволяет отправлять на сервер более сложные данные.

````

````python
from flask import request

app = Flask(__name__)


@app.route("/api/sum", methods=["POST"])
def api_sum():
    try:
        data = request.get_json()
        number1 = data.get("number1")
        number2 = data.get("number2")
        result = number1 + number2
        return jsonify({"result": result, "status": "успех"})
    except Exception as e:
        return jsonify({"error": str(e), "status": "ошибка"}), 400
    
run_simple('localhost', 5059, app, use_reloader=False)
````

````python
from flask import Flask, render_template, request, redirect, url_for
import os
from openpyxl import load_workbook

app = Flask(__name__)

# Default file name for the schedule
schedule_file = "schedule.xlsx"
schedule = {}

# Function to load schedule from the Excel file
def load_schedule(file_path):
    global schedule
    try:
        workbook = load_workbook(file_path)
        sheet = workbook.active
        headers = [sheet.cell(row=7, column=col).value for col in range(1, sheet.max_column + 1)]

        schedule = {}  # Reset schedule
        for row in sheet.iter_rows(min_row=8, values_only=True):
            day = row[1]  # День недели (столбец B)
            date = row[2]  # Дата (столбец C)
            time = row[4]  # Время (столбец E)
            if not date or not time:
                continue

            if "обед" in str(row[5]).lower():  # Пропуск строк с "Обед"
                continue

            for col_idx, group_name in enumerate(headers[5:], start=6):  # Начинаем с группы
                if group_name:
                    if group_name not in schedule:
                        schedule[group_name] = []
                    schedule[group_name].append(
                        {"day": day, "date": date, "time": time, "lesson": row[col_idx]}
                    )
    except Exception as e:
        print(f"Ошибка загрузки расписания: {e}")

# Load the default schedule on startup
if os.path.exists(schedule_file):
    load_schedule(schedule_file)

# Default route to show the current schedule
@app.route("/")
def index():
    return render_template("index.html", schedule=schedule)

# Route to handle file uploads
@app.route("/upload", methods=["GET", "POST"])
def upload():
    if request.method == "POST":
        if "file" not in request.files:
            return "No file part"
        file = request.files["file"]
        if file.filename == "":
            return "No selected file"
        if file:
            file.save(schedule_file)
            load_schedule(schedule_file)
            return redirect(url_for("index"))
    return render_template("upload.html")

# Run the application
if __name__ == "__main__":
    app.run(debug=True)
````
